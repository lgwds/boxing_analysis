```python
import c3d
import pandas as pd
import numpy as np
```

```python
data = []
with open('DeviceAbduction01.c3d', 'rb') as handle:
    reader = c3d.Reader(handle)
    for i, points, analog in reader.read_frames():
        #print('frame {}: point {}, analog {}'.format(i, points.shape, analog.shape))
        data.append(points)
        
point_labels = np.array(reader.point_labels)
```

```python
arr = np.array(data)
lable_two = np.array(['x', 'y', 'z', 'res', 'res2'])
cols = pd.MultiIndex.from_product([point_labels, lable_two])
```

```python
arr = np.array(data)
#df = pd.DataFrame(data=arr[:,:,0],columns=point_labels)
df = pd.DataFrame(data=arr.T.reshape(arr.shape[0],-1), columns=cols)
df
```
